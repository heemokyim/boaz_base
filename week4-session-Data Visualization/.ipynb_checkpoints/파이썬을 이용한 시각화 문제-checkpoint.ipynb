## df.plot.barh()를 이용하여 total_bedrooms population household의 ocean_proximity 별 평균을 시각화해보아라

```python
## 이 다섯개 ocean_proximity 에 따른 
#housing_median_age, total_rooms, total_bedrooms, population, households, median_income을 알고 싶다면?
## 각각 평균을 내고 ocean_proximity 에 따라 bar plot을 그리면 된다. 
## (1) 먼저 이에 해당되는 dataframe 형성(groupby)
## 이때 total rooms 와 median_house_value housing_median_age는 단위가 혼자 튀기 때문에 
#이 두변수를 제거한 dataframe을 만들어줍니다. 


col = cali[['total_bedrooms','population', 'households','ocean_proximity']]
group = col.groupby(['ocean_proximity'])
group.mean().plot.barh()
```

## Subplot을 이용하여 4개의 그래프를 만들어라! 

```python
## 변수들 모두 subplot을 한 눈에 알아볼 수 있게 배열 할 수 있다. 
## 4개의 dataframe형성 

col2 = cali[['total_bedrooms','population','total_rooms', 'households','ocean_proximity']]
mean_by_op = col2.groupby(['ocean_proximity']).mean()
mean_by_op
```

```python

color = ['b', 'g', 'r', 'c', 'm']

x_=mean_by_op.index


plt.subplot(3,3,1)
plt.barh(x_,mean_by_op['total_bedrooms'],0.5,color=color)
plt.title('total_bedrooms')
plt.ylabel('ocean_proximity')

plt.subplot(3,3,3)
plt.barh(x_,mean_by_op['population'],0.5,color=color)
plt.title('population')
plt.ylabel('ocean_proximity')

plt.subplot(3,3,7)
plt.barh(x_,mean_by_op['households'],0.5,color=color)
plt.title('households')
plt.ylabel('ocean_proximity')

plt.subplot(3,3,9)
plt.barh(x_,mean_by_op['total_rooms'],0.5,color=color)
plt.title('total_rooms')
plt.ylabel('ocean_proximity')
```

```python
## 그 중에서 4개의 변수만 그려본다
##subplot할때 nrow =3, ncol = 3 해야 이쁘게 나온다.


```

## Median_income 과 Median_house_value의 Scatterplot을 그려보자

```python
plt.title('income and house_value scatterplot')
plt.scatter(cali['median_income'],cali['median_house_value'])
plt.xlabel('median_income')
plt.ylabel('median_house_value')
```

```python
##색깔을 입혀보자. 거주지역이 NEAR_BAY인 사람들과 아닌 사람들로 나눠서 보고싶다면?
cols = []
for i in cali['ocean_proximity']:
    choice = 0
    if i == 'NEAR OCEAN':
        choice = 1
        
    else:
        choice = 2
    cols.append(choice)
```

```python
##방금만든 cols를 활용한다. 점들이 겹치기 때문에 투명도도 준다.(c, alpha 활용)

plt.scatter(cali['median_income'],cali['median_house_value'], alpha=0.2,c=cols)
plt.xlabel('median_income')
plt.ylabel('median_house_value')
```

## 위도와 경도 자료를 활용, size는 population/100, color를 median_house value로 해서 지도를 그려보아라

```python
plt.scatter(cali['longitude']
            ,cali['latitude']
            ,c = cali['median_house_value']
            ,s = cali['population']/100
            ,cmap='Greys')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
```
